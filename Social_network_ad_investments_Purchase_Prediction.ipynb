# Logistic Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Datasets\Social_Network_Ads.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
plt.figure(figsize=(20,15))
plt.plot(dataset['EstimatedSalary'],dataset['Purchased'])
dataset.columns
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Without scaling features, the algorithm may be biased towards the feature which has values higher in magnitude. This makes training easy. It prevents the optimization from getting stuck in local optima.

```python
print(X_train)
```

```python
print(X_test)
```

## Training the Logistic Regression model on the Training set

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

## Predicting a new result

```python
#From the classifier object we use the predict method,to find the purchase decision of aged 30 yrs with salary 87000
#The predict method can only work on scaled data,as the training data was scaled
```

```python
print(classifier.predict(sc.transform([[30,87000]])))   #1st observation in X-test #we use [][] as we expect a 1*2 matrix
```

## Predicting the Test set results

```python
#Concatenate function of numpy allows us to combine two vectors horizontally or vertically. In this case as we will print 
#the pred and actual values vertically,we will go for the reshape function which will help us with the vertical display
#len(y_pred) will obviously be 1,and an extra 1 means(1 column).similarly for the actual values in y_test we will have the 
# unit value displayed as a column input. 
#The output is a tuple with two reshaped values.
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
